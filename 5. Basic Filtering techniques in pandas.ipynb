```python
import pandas as pd
```

```python
orders = pd.read_excel("./Sample - Superstore Sales.xls",0)
```

```python
orders.head(2)
```

### Filtering the data based on numbers

```python
orders[orders.Sales >=500]
```

### Flitering the data based on the String / Character :

```python
orders[orders["Order Priority"]!= "High"]
```

### Filtering the data with mutliple factors :

```python
orders[(orders['Order Priority'] !='High')& (orders['Customer Segment']=='Corporate')]
```

```python
orders[(orders['Order Priority'] !='High') | (orders['Customer Segment']=='Corporate')]
```

### Shorten way of implementing the OR condition:

```python
orders[orders["Ship Mode"].isin([
    "Regular Air","Delivery Truck"
])]
```
